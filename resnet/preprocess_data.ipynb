```python
df = pd.read_csv("./lidar_data_with_raw_seconds.csv")
df
```

```python
data = df.copy()
```

```python
data["filename"] = "None"
data["train"] = "True"
data
```

```python
y, sr = librosa.load("./audio_22092022.WAV", sr=None)
```

```python
for i, row in data.iterrows():
    entry, exit = row["entry"], row["exit"]
    if i / len(df) < 0.8:
        filename = "train_{}.wav".format(i)
        wavpath = "./train/train_{}.wav".format(i)
    else:
        filename = "valid_{}.wav".format(i)
        wavpath = "./valid/valid_{}.wav".format(i)
        data.loc[i, "train"] = "False"
    data.loc[i, "filename"] = filename
    #data.loc[i, "target"] = category_to_label[clas]
    start, end = int((entry - 1)*sr), int((exit + 1)*sr)
    if start < 0: start = 0
    if end > len(y): end = len(y)-1
    assert start > 0
    assert end < len(y)
    #start, end = int((entry * sr), int(exit * sr)
    sf.write(wavpath, y[start:end], sr, 'PCM_24')  # slice by seconds, seconds in audio = SR * seconds
```
